## Heart Disease Dataset - Multivariate Analysis

#### Objective

To identify correlations between more than two parameters from the heart disease dataset, so that we can identify a heart disease patient more easily and efficiently.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"clean_heart_data.csv")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='age', hue='heart_disease', data=df)
plt.title('Age vs Gender vs Heart Disease')
plt.xlabel('Gender (0=Female, 1=Male)')
plt.ylabel('Age')
plt.legend(title='Heart Disease')
```

For both male and female patients, those less than age 50 are comparatively at a lower risk of heart disease than patients aged>50. For females, the maximum number of patients above age 70 is identified with heart disease, whereas for males, the data is mixed in this category. Both of the groups cluster around 60, revealing a hotspot of heart disease.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='chest_pain_type', y='age', hue='heart_disease', data=df)
plt.title('Age vs Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Category')
plt.ylabel('Age')
plt.legend(title='Heart Disease')
```

```python
pd.crosstab(df['chest_pain_type'], df['heart_disease'])
```

The relationship between age, chest pain type, and heart disease shows significant overlap across all groups. While certain chest pain types (such as cp = 1 and cp = 3) show slightly higher median ages for patients with heart disease, the distributions are largely overlapping. This indicates that age alone does not strongly differentiate the presence of heart disease within chest pain categories, and other factors are likely involved.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='cholesterol_mg/dl',
    y='max_heart_rate',
    hue='heart_disease',
    data=df,
    alpha=0.6
)
plt.title('Cholesterol vs Max Heart Rate by Heart Disease')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.legend(title='Heart Disease (0=No, 1=Yes)')
plt.show()
```

* The data is highly complex.
* We see a dense cluster of orange dots at around cholesterol level 200mg/dl and Max Heart Rate 125. This implies that patients having stats around these values are more likely to have heart disease.
* Around the point of cholesterol level 125 mg/dl and max heart rate about 165, same sort of orange dot cluster is seen.
* Some cholesterol level outliers are notable. MAximum of patients having these values are non-disease cases.
* Most of the data ranges from cholesterol level 100-300 and heart rate 100-200. This is our core population zone.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='resting_bp_mmhg',
    y='st_depression',
    hue='heart_disease',
    data=df,
    alpha=0.6
)
plt.title('ST Depression vs Resting BP by Heart Disease')
plt.xlabel('Resting BP')
plt.ylabel('ST depression')
plt.legend(title='Heart Disease (0=No, 1=Yes)')
```

* The data is highly spread. So, no clear identifier data can be provided.
* The core data lies between- Blood Pressure (100-150) and ST depression (0-2).
* A weak pattern can be identified as at st depression 2-4 there are more orange dots, meaning this data shows a higher possibility of heart disease.
* Overall, these features do not act as strong standalone predictors and should be considered alongside other variables.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='cholesterol_mg/dl',
    y='max_heart_rate',
    hue='chest_pain_type',
    data=df,
    alpha=0.6,
    palette='Set1'
)
plt.title('Cholesterol vs Max Heart Rate by Chest Pain type')
plt.xlabel('Cholesterol')
plt.ylabel('Max Heart Rate')
plt.legend(title='Chest Pain Type')
```

The scatter plot of cholesterol and maximum heart rate across different chest pain types shows a high degree of overlap between all categories. No distinct clustering is observed, indicating that chest pain types are distributed across similar ranges of cholesterol and heart rate. Most observations fall within moderate ranges, where all chest pain categories coexist. Additionally, a few extreme cholesterol values are present, likely representing outliers. Overall, the chest pain type does not show a strong relationship with these continuous variables in isolation.

```python
df.corr()['heart_disease'].sort_values(ascending=False)
```

No parameter has a strong correlation with heart disease. It is expected as heart disease are highly complex will need to very different data accurately to confirm a heart disease.

## Key Insights:

1. Patients aged 60 are more vulnerable to heart diseases than any other age group.
2. Age alone is not a strong differentiator among chest pain groups for heart disease.
3. ST depression/Max Heart Rate/Cholesterol level, any of these is not alone a powerful parameter in the identification of heart disease.

## Summary

Heart diseases are super complex, and they depend upon many parameters at the same time. So identifying heart disease is as hard as solving a hard puzzle. Though we have analyzed around 1000 patient data and luckily found some patterns that really give us strong insights that will probably be very helpful to doctors for the purpose of identifying heart diseases.
