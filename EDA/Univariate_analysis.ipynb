## Heart Disease Dataset - Univariate Analysis

Objective :To understand the distribution, spread, and basic characteristics of each feature in the dataset.

```python
import pandas as pd
df = pd.read_csv(r"clean_heart_data.csv")
df.head()
```

```python
df.describe()
```

1. Mean age is 60, patients are mostly around age of 60s.
2. Chest pain type varies highly as std=1.46 from mean=0.97. Patients have different types of chest pains.
3. Mean BP of patients = 121.73, with std=18.20 meaning patients have this range of Blood pressure(110-130).
4. Patients have a mean heart rate of approx. 150, this is a very strong indicator of heart diseases.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['age'],bins=20,kde=True)
plt.title('Age-wise cases')
```

It is evident that age ranges vary highly from 30-90. However, around 60, the count is highest.

It means people around age of 60s are more vulnerable to heart diseases.

```python
sns.countplot(x=df['gender'])
plt.title('Gender comparison')
plt.xlabel('gender: 1=Male 0=Female')
```

As it is seen that there are around 300 female(0) data and rest 500 data refers for male(1). In fact, male are more vulnerable to heart diseases.

```python
sns.countplot(x=df['chest_pain_type'])
plt.title('chest pain types')
```

Chest pain type varies and almost the same count for each category. This means that heart patients may have several types of chest pain or no chest pain at all(-1).

```python
sns.histplot(df['resting_bp_mmhg'],kde=True,bins=20)
plt.title('Resting Blood Pressure range')
```

It is prevalant that resting blood pressure ranges from 110-140 for patients. however the graph is right skewed. so patients may have a high blood pressure(>150) as a symptomp of heart disease.

```python
sns.histplot(df['cholesterol_mg/dl'],kde=True,bins=20)
plt.title('Patient Cholesterol level')
```

Most of the patients cholesterol level varies from 150-300. However outliers are also seen(>400). This means that patients may also have higher values tending to Heart Disease.

```python
sns.histplot(df['max_heart_rate'],kde=True,bins=20)
plt.title('Patient Heart Rate Count')
```

The heart rate varies highly however most of the cases are from 130-180 meaning this is a typical value of heart rate for a patient with heart problems. The graph is right skewed meaning heart rate values may also be abnormally high indicating high chance of heart disease.

```python
sns.histplot(df['st_depression'],kde=True,bins=20)
plt.title('Patient ST depression')
```

It is evident that most cases ranges from 0-2 meaning that there are as many cases of heart issues as normal. The range also goes up to 3-4 meaning many patients have serious heart problems.

#### N.B. ST depression range:
#### 0-1 Normal
#### 1-2 Mild Abnormality
#### 3-4 Strong indicator of heart disease
